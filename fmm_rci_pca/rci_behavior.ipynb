```python
import pandas as pd
import pandas_ta as ta
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/projetofinal_ia/fmm_rci_pca/database/BTCUSDT3600.csv')
data['date'] = data['date'].astype('datetime64[s]')
data = data.set_index('date')
```

```python
rsi_periods = list(range(2, 25))

rsis = pd.DataFrame()
for p in rsi_periods:
    rsis[p] = ta.rsi(data['close'], p)
```

```python
plt.figure(figsize=(18, 12))
rsis.hist(bins=100)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)  # Ajustando o espaçamento entre gráficos

# Tamanhos de fontes menores
plt.xlabel("RSI Value", fontsize=10)
plt.ylabel("Frequency", fontsize=10)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

# Mostrando o histograma
plt.show()

# Aumentando a figura do heatmap em 50%
plt.figure(figsize=(15, 12))
sns.heatmap(rsis.corr(), annot=True, cmap="coolwarm", fmt=".2f")

# Tamanhos de fontes menores
plt.xlabel("RSI Period", fontsize=10)
plt.ylabel("RSI Period", fontsize=10)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

# Mostrando o heatmap
plt.show()
```
